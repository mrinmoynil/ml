## find how many people use ' python ' in a country

###  using filtering 

```python
filt=df['Country']=='India'
df.loc[filt]['LanguageWorkedWith']
```

```python
df.loc[filt]['LanguageWorkedWith'].str.contains('Python')
```

```python
df.loc[filt]['LanguageWorkedWith'].str.contains('Python').sum()
```

###  using groupby

```python
#it shows errror because it not a series, it is a series groupby object

country_grp['LanguageWorkedWith'].str.contains('Python').sum()
```

###  to remove this problem ,we need to use  apply function

```python
# here ' x ' is a series


country_grp['LanguageWorkedWith'].apply(lambda x: x.str.contains('Python').sum())
```
